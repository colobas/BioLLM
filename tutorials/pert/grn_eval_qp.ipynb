```python
from biollm.task.grn.network_task import GrnTask
from biollm.evaluate.bm_metrices_grn import evaluate

task1 = GrnTask(cfs_file="../config/scgpt_grn.toml")

i = 0.2
g1_eval = []
while i < 0.7:
    i += 0.1
    g1 = task1.grn_analysis(threshold=i)
    eval_res = evaluate(g1, modularity = True)
    g1_eval.append(eval_res)
```

```python
from biollm.task.grn.network_task import GrnTask
from biollm.evaluate.bm_metrices_grn import evaluate

task2 = GrnTask(cfs_file="../config/scmamba_grn.toml")

i = 0.2
g2_eval = []
while i < 0.7:
    i += 0.1
    g2 = task2.grn_analysis(threshold=i)
    eval_res = evaluate(g2, modularity = True)
    g2_eval.append(eval_res)
```

```python
import matplotlib.pyplot as plt

#  Modularity  Biological Index 
modularity_g1 = [eval_res['Modularity'] for eval_res in g1_eval]
biological_index_g1 = [eval_res['Biological Index'] for eval_res in g1_eval]
modularity_g2 = [eval_res['Modularity'] for eval_res in g2_eval]
biological_index_g2 = [eval_res['Biological Index'] for eval_res in g2_eval]

x_value = [0.2, 0.3, 0.4, 0.5]
# 
plt.figure(figsize=(14, 5))
# 1 Modularity
plt.subplot(1, 2, 1)
plt.plot(x_value, modularity_g1, label='scGPT Modularity', marker='o')
plt.plot(x_value, modularity_g2, label='scMamba Modularity', marker='x')
plt.xlabel('Cosine Similarity Cutoff')
plt.ylabel('Modularity Value')
plt.title('Comparison of Modularity between g1 and g2')
plt.legend()
plt.grid(True)
# 2 Biological Index
plt.subplot(1, 2, 2)
plt.plot(x_value, biological_index_g1, label='scGPT Biological Index', marker='o')
plt.plot(x_value, biological_index_g2, label='scMamba Biological Index', marker='x')
plt.xlabel('Cosine Similarity Cutoff')
plt.ylabel('Biological Index Value')
plt.title('Comparison of Biological Index between g1 and g2')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

def plot_grouped_line_chart(x, y_groups, group_labels):
    # Create a plot
    plt.figure(figsize=(8, 6))
    
    # Draw the line chart
    for i, y in enumerate(y_groups):
        plt.plot(x, y, marker='o', label=group_labels[i])
    
    # Add title and labels
    plt.title('Grouped Line Chart', fontsize=16)
    plt.xlabel('X-axis', fontsize=12)
    plt.ylabel('Y-axis', fontsize=12)
    
    # Add a legend
    plt.legend()
    
    # Show the plot
    plt.show()

# Sample data
x = [1, 2, 3, 4, 5]
y1 = [1, 2, 3, 4, 5]
y2 = [2, 3, 4, 5, 6]
group_labels = ['Group A', 'Group B']

# Call the function to plot the grouped line chart
plot_grouped_line_chart(x, [y1, y2], group_labels)
```
